```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("train (1).csv")
```

```python
df.head()
```

# 1. Categorical Data

<h1>a.countplot</h1>

```python
# Get the order based on value counts
order = df['Survived'].value_counts().index

# Seaborn countplot (uses same order)
sns.countplot(x='Survived', data=df, order=order, palette='rocket')

# Pandas bar plot (same order + same colors)
df['Survived'].value_counts()[order].plot(kind='bar', color=sns.color_palette('rocket'))

plt.show()
```

```python
# Get the order based on value counts
order = df['Pclass'].value_counts().index

# Seaborn countplot (uses same order)
sns.countplot(x='Pclass', data=df, order=order, palette='icefire')

# Pandas bar plot (same order + same colors)
df['Pclass'].value_counts()[order].plot(kind='bar', color=sns.color_palette('icefire'))

plt.show()
```

```python
# Get the order based on value counts
order = df['Sex'].value_counts().index

# Seaborn countplot (uses same order)
sns.countplot(x='Sex', data=df, order=order, palette='twilight')

# Pandas bar plot (same order + same colors)
df['Sex'].value_counts()[order].plot(kind='bar', color=sns.color_palette('twilight'))

plt.show()
```

```python
# Get the order based on value counts
order = df['Embarked'].value_counts().index

# Seaborn countplot (uses same order)
sns.countplot(x='Embarked', data=df, order=order, palette='pastel')

# Pandas bar plot (same order + same colors)
df['Embarked'].value_counts()[order].plot(kind='bar', color=sns.color_palette('pastel'))

plt.show()
```
