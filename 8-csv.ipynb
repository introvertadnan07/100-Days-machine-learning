# 3. Opening a csv file from an URL

```python
import pandas as pd
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/dr5hn/countries-states-cities-database/master/csv/countries.csv"

response = requests.get(url)
response.raise_for_status()  # ensures it throws an error if the download fails

data = StringIO(response.text)
df = pd.read_csv(data)

print(df.head())
```

# 14. Convertors

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return 'RCB'
    else:
        return name

print(rename("Royal Challengers Bangalore"))
```

```python
pd.read_csv("matches.csv", converters={'team1': rename})
```
