```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
```

```python
%matplotlib inline
plt.scatter(X,y)
plt.show()
```

```python
lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
import time

b = 150
m = -127.82
all_b = []
all_m = []
all_cost = []
all_lr = []


epochs = 1

start = time.time()

t0,t1 = 5,50
def learning_rate(t):
    return t0/(t + t1)

for i in range(epochs):
    for j in range(X.shape[0]):
        
        lr = learning_rate(i * X.shape[0] + j)
        
        idx = np.random.randint(X.shape[0], size=1)
        
        slope_b = -2 * (y[idx] - (m * X[idx]) -  b)
        slope_m = -2 * (y[idx] - (m * X[idx]) -  b) * X[idx]
        cost = (y[idx] - m * X[idx] -b) ** 2
        
        
        b = b - (lr * slope_b)
        m = m - (lr * slope_m)
        all_b.append(b)
        all_m.append(m)
        all_cost.append(cost)
        all_lr.append(lr)
print("Total time taken",time.time() - start)
```

```python
fig, ax = plt.subplots(figsize=(9,5))
#fig.set_tight_layout(True)

x_i = np.arange(-3, 3, 0.1)
y_i = x_i*(-27) -150
ax.scatter(X, y)
line, = ax.plot(x_i, x_i*50 - 4, 'r-', linewidth=2)

def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i*all_m[i] + all_b[i])
    ax.set_xlabel(label)
    # return line, ax

anim = FuncAnimation(fig, update, frames=100, interval=5)

f = r"stochastic_animation_line_plot.gif" 
writergif = animation.PillowWriter(fps=2) 
anim.save(f, writer=writergif)
```

```python
# Use interactive Matplotlib inside Jupyter
%matplotlib widget

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation

# ----------------------------------------------------------
# Create synthetic data for demonstration
# ----------------------------------------------------------

# Create a grid of slope (m) and intercept (b) values
m_arr = np.linspace(-150, 150, 100)
b_arr = np.linspace(-150, 150, 100)

# Create meshgrid for contour plot
M, B = np.meshgrid(m_arr, b_arr)

# Define a dummy "loss surface" (you can replace it with your real one)
z_arr = (M - 30)**2 + (B - 50)**2  # example cost surface

# Simulated learning path (values of m and b per epoch)
num_epochs = 30
all_m = np.linspace(-100, 80, num_epochs)
all_b = np.linspace(-120, 60, num_epochs)

# ----------------------------------------------------------
# Plot setup
# ----------------------------------------------------------
fig = plt.figure(figsize=(9, 5))
axis = plt.axes(xlim=(-150, 150), ylim=(-150, 150))

# Contour background
contour = axis.contourf(m_arr, b_arr, z_arr, cmap='viridis')

# Line for the animation
line, = axis.plot([], [], lw=2, color='white')
xdata, ydata = [], []

# ----------------------------------------------------------
# Animation function
# ----------------------------------------------------------
def animate(i):
    label = f"epoch {i + 1}"
    xdata.append(all_m[i])
    ydata.append(all_b[i])
    line.set_data(xdata, ydata)
    axis.set_xlabel(label)
    return line,

# ----------------------------------------------------------
# Run animation
# ----------------------------------------------------------
anim = FuncAnimation(fig, animate, frames=num_epochs, 
                     repeat=False, interval=200)

plt.show()
```
