# 2. Bar Plot(Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(data = titanic, x = 'Pclass', y = 'Age', palette='rainbow')
plt.show()
```

```python
sns.barplot(data = titanic, x = 'Pclass', y = 'Fare', palette='rainbow', hue='Sex')
plt.show()
```

```python
sns.barplot(data = titanic, x = 'Pclass', y = 'Age', palette='rainbow', hue='Sex')
plt.show()
```

# 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(data = titanic, 
            x = 'Sex', 
            y = 'Age', 
            hue = 'Survived')
plt.show()
```

# 4. Displot (Numerical - Categorical)

```python
sns.displot(titanic[titanic['Survived'] == 0]['Age'], kde = True)
plt.show()
```

```python
sns.kdeplot(data=titanic[titanic['Survived'] == 0]['Age'], label='Did not survive')
sns.kdeplot(data=titanic[titanic['Survived'] == 1]['Age'], label='Survived')
plt.legend()
plt.show()
```

# 5. HeatMap (Categorical - Categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
crosstab = pd.crosstab(titanic['Pclass'], titanic['Survived'], margins=True)
sns.heatmap(crosstab, annot=True, fmt='d', cmap='YlGnBu')
plt.show()
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind = 'bar')
```

```python
titanic.groupby('Sex').agg({'Survived': 'mean'}) * 100
```

# 6. ClusterMap(Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived']))
```

# 7. Pairplot

# 8. Lineplot(Numerical - Numerical)

```python
print(flights.columns)
```

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
```

```python
new
```

```python
sns.lineplot(x='year', y='passengers', data=new)
plt.show()
```

```python
sns.heatmap(
    flights.pivot_table(values='passengers', index='month', columns='year'),
    annot=True,
    cmap='YlGnBu'
)
plt.show()
```

```python
sns.clustermap(
    flights.pivot_table(values='passengers', index='month', columns='year')
)
plt.show()
```
